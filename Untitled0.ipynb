```python
import numpy as np
import pandas as pd
```

```python
#importação do arquivo
df=pd.read_csv('agricultural_raw_material.csv')
```

Explorando conjunto de dados

```python
df.info
#checando se tem valores NULL
df.isnull().sum()
```

Tratando com dados ausentes, incorretos e invalidos

```python
df = df.replace('%','', regex=True)
df = df.replace(',','', regex=True)
df = df.replace('-','', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

df
```

```python
#apagando valores nulos

df = df.dropna()
df
```

```python
#retornando True onde há NaN, e False onde há valores válidos.
df.isnull().sum()
```

```python
#convertendo tipos de dados pra float
lst =["Coarse wool Price","Coarse wool price % Change","Copra Price","Copra price % Change","Cotton Price","Cotton price % Change","Fine wool Price","Fine wool price % Change","Hard log Price","Hard log price % Change","Hard sawnwood Price","Hard sawnwood price % Change","Hide Price","Hide price % change","Plywood Price","Plywood price % Change","Rubber Price","Rubber price % Change","Softlog Price","Softlog price % Change","Soft sawnwood Price","Soft sawnwood price % Change","Wood pulp Price","Wood pulp price % Change"]
df[lst] = df[lst].astype("float")
df.dtypes
```

```python
#trazendo Top 5 no nosso dataframe
df.head()
```

Tratar a coluna Month (transformar ela em formato de data)

```python
#usaremos a função pd.to_datetime que faz essa conversão
print(df['Month'].dtype)

#df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
#indexando o mes

```

Criando paineis

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

Mapa de calor

```python
raw_data = ['Coarse wool Price','Copra Price','Cotton Price','Fine wool Price','Hard log Price','Hard sawnwood Price',
            'Hide Price','Plywood Price','Rubber Price','Softlog Price','Soft sawnwood Price','Wood pulp Price']
corrmat = df[raw_data].corr()

fig = plt.figure(figsize=(12, 9))

mask = np.triu(np.ones_like(corrmat, dtype=bool))

sns.heatmap(corrmat, vmax= .8,mask=mask, square = True, annot = True)

plt.show()
```

O mapa de calor acima mostra o valor correlacionado entre as materias-primas, e quanto mais proximo de 1 mais correlacionada estão.

```python
plt.figure(figsize=(15,5))
changeList=["Coarse wool price % Change","Copra price % Change","Cotton price % Change","Fine wool price % Change","Hard log price % Change","Hard sawnwood price % Change","Hide price % change",
            "Plywood price % Change","Rubber price % Change","Softlog price % Change","Soft sawnwood price % Change","Wood pulp price % Change"]
corrMatrix = df[changeList].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

Quase não há variação entre os porcentuais e o valor

o valor negativo indica que as duas variaveis estão negativamente
correlacionadas


0 é nenhuma relação


Quanto maior o valor, maior a correlação

Agora vou responder perguntas de negócio

```python
changeList=["Coarse wool price % Change","Copra price % Change","Cotton price % Change","Fine wool price % Change","Hard log price % Change","Hard sawnwood price % Change","Hide price % change",
            "Plywood price % Change","Rubber price % Change","Softlog price % Change","Soft sawnwood price % Change","Wood pulp price % Change"]
for i in range(len(changeList)):
  plt.figure(figsize=(12,12))
  df[changeList[i]].hist(figsize=(11,9), linewidth=1)
  plt.xlabel('% Change')
  plt.ylabel('count')
  plt.legend(changeList[i:],loc='upper center', bbox_to_anchor=(1.2,1))
```

Descobrindo variações dos preços

```python
plt.figure(figsize=(10,10))
materialList= ['Coarse wool Price','Copra Price','Cotton Price','Fine wool Price','Hard log Price','Hard sawnwood Price',
            'Hide Price','Plywood Price','Rubber Price','Softlog Price','Soft sawnwood Price','Wood pulp Price']
for i in range(len(materialList)):
  plt.subplot(4,3,i+1)
  plt.subplots_adjust(hspace=1, wspace=0.5)
  plt.title(materialList[i])
  plt.plot(df[materialList[i]])
  plt.xticks(rotation=90)
plt.suptitle('Price comparation')
```

Comparando os precos e podemos ver que o cotton e o rubber são os menores

```python
plt.figure(figsize=(10,10))
plt.plot(df[['Cotton Price','Rubber Price']])
plt.title('Comparação dos dois menores preços')
plt.xlabel('anos')
plt.ylabel('Preços')
plt.legend(['Cotton Price','Rubber Price'], loc='upper center', bbox_to_anchor=(1.2,1))
```

Podemos comparar o preço dos dois menores valores, e dizer quem esta valendo mais antigamente e nos tempo atuais.
